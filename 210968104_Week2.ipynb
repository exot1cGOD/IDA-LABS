```python
import pandas as pd
```

```python
data=[]
for i in range(10):
    roll=int(input("Enter your roll no: "))
    name=input("Enter your name: ")
    gender=input("Enter your gender: ")
    mark1=int(input("Enter marks 1: "))
    mark2=int(input("Enter marks 2: "))
    mark3=int(input("Enter marks 3: "))
    data.append([roll,name,gender,mark1,mark2,mark3])
```

```python
df=pd.DataFrame(data,columns=["Roll Number","Name","Gender","Marks1","Marks2","Marks3"])
```

```python
df
```

```python
dflist=["Marks1","Marks2","Marks3"]
df["Total Marks"]= df[dflist].sum(axis=1)
```

```python
df
```

```python
low_mark1=df['Marks1'].min(axis=0)
low_mark1
```

```python
high_mark2=df['Marks2'].max(axis=0)
high_mark2
```

```python
avg_mark3=df['Marks3'].mean(axis=0)
avg_mark3
```

```python
df["Average Marks"]= df[dflist].mean(axis=1)
high_avg=df[df["Average Marks"]==df['Average Marks'].max(axis=0)]
high_avg
```
